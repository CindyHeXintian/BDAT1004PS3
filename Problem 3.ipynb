```python
#4
import pandas as pd
wind = pd.read_csv('wind.txt', delimiter='\s+')
wind
```

```python
wind.rename(columns={'Yr':'Year','Mo':"Month","Dy":"Day"}, inplace=True)
wind
```

```python
wind['Year'] += 1900
wind
```

```python
wind.index = pd.to_datetime(pd.concat([wind['Year'], wind['Month'], wind['Day']], axis=1))
wind.drop(['Year','Month','Day'], axis=1, inplace=True)
wind
```

```python
wind.isnull().sum().sum()
```

```python
wind.notnull().sum().sum()
```

```python
wind.mean().mean()
```

```python
wind[:24]
```

```python
loc_stats = pd.concat([wind.mean(), wind.min(), wind.max(), wind.std()], axis=1)
loc_stats.rename(columns={0:'Mean',1:'Min',2:'Max',3:'Std'}, inplace=True)
loc_stats
```

```python
day_stats = pd.concat([wind.mean(axis=1), wind.min(axis=1), wind.max(axis=1), wind.std(axis=1)], axis=1)
day_stats.rename(columns={0:'Mean',1:'Min',2:'Max',3:'Std'}, inplace=True)
day_stats
```

```python
jan = wind[wind.index.month == 1]
jan.mean()
```

```python
yearly = wind.resample('Y').ffill()
yearly
```

```python
monthly = wind.resample('M').ffill()
monthly
```

```python
weekly = wind.resample('W').ffill()
weekly
```

```python
weekly_stats = weekly[:52]
weekly_stats = pd.concat([weekly_stats.mean(), weekly_stats.max(), weekly_stats.mean(), weekly_stats.std()], axis=1)
weekly_stats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)
weekly_stats
```

```python
#6
us = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us.index = us['Year']
mrg_per_capita = us['Marriages_per_1000'] / 1000
div_per_capita = us['Divorces_per_1000'] / 1000

df = pd.concat([mrg_per_capita, div_per_capita], axis=1).rename(columns={'Marriages_per_1000':'Marriages_per_capita','Divorces_per_1000':'Divorces_per_capita'})
mask = (df.index >= 1867) & (df.index <= 2014)
df.loc[mask].plot(xlabel='Year', ylabel='Marriages/Divorces per capita')
```

```python
#7
mask = (df.index == 1900) | (df.index == 1950) | (df.index == 2000)
df.loc[mask].plot.bar(xlabel='Year', ylabel='Marriages/Divorces per capita')
```

```python
#9
roman = pd.read_csv('roman-emperor-reigns.csv')
mask = roman['Cause_of_Death'].str.contains('ssassin')
assassinated = roman.loc[mask].set_index('Emperor').drop(['Length_of_Reign'], axis=1)
assassinated.groupby('Cause_of_Death').size().plot.pie(ylabel='Number of Emperors')
```
